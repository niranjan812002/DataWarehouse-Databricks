```sql
%sql
CREATE OR REPLACE VIEW sales_scd.view_DimProducts
AS
SELECT DISTINCT(ProductID) as ProductID, ProductName, ProductCategory
FROM sales_scd.orders
where OrderDate>"2024-02-10"
```
